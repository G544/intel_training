```python
a = [int(s) for s in input().split(' ')]

d = []
count = 0
for item in a:
    if item == 2:
        if count == 0:
            d.append([s for s in range(10)])
        elif count == 10:
            d.append([s for s in range(10, 0, -1)])
        else:
            tmp1 = [s for s in range(10-count)]
            tmp2 = [s for s in range(count,0, -1)]
            d.append(tmp2+tmp1)
    count+=1
count = 0 
mas_dist = []
for item in a:
    if item == 1:
        tmp = []
        for i in d:
            tmp.append(i[count])
        mas_dist.append(min(tmp))
    count += 1
print(max(mas_dist))
        
```

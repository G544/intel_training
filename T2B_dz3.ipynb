```python
a = ([(s) for s in input().split(' ')])
b = {}
for item in a:
    if item in b:
        b[item] = b[item]+1
    else:
        b[item] = 1

c = {key:val for key,val in b.items() if val==1}
print(' '.join(list(c.keys())))
```

# Задача D

```python
n = int(input())
s=input().strip()
possible = set(range(1,n+1))
while s != 'HELP':
    nums = set(map(int, s.split()))
    s = input()
    if s == 'YES':
        possible.intersection_update(nums)
    elif s=='NO':
         possible.difference_update(nums)
    s = input().strip()
    
print(*sorted(possible))
```

```python
num_sv = int(input())
my_list =[]
for i in range(num_sv):
    my_list.append(set(input()))
n_nums = int(input())
analys_dict = {}
#max_val = 0
#num_max = ''
for i in range(n_nums):
    num_ich = input()
    num = set(num_ich)
    count = 0
    for item in my_list:
        if item.issubset(num):
            count += 1
    #if count > max_val:
    #    max_val = count
    #    num_max = num_ich
    analys_dict[num_ich] = count

max_val = max(analys_dict.values())
num_max = [key for key,val in analys_dict.items() if val == max_val]
        
for item in (num_max):
    print(item)
```

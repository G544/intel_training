```python
parties = {}
c_all = 0

with open('input.txt', 'r') as fin:
    for s in fin:
        parties[' '.join(s.split()[:-1])] = int(s.split()[-1])
        c_all += int(s.split()[-1])
        


izb_ch = c_all/450
first = {}
second = []
for key, val in parties.items():
    first[key] = val//izb_ch
    second.append([key, val % izb_ch, val])
second = (sorted(second, key=lambda x: (x[1], x[2]), reverse = True))
count = 0
while sum(first.values()) < 450:
    first[second[count][0]] = first[second[count][0]]+1
    count += 1
    if count == len(second):
        count = 0
        
for key, value in first.items():
      print("{0} {1}".format(key,int(value)))
```
